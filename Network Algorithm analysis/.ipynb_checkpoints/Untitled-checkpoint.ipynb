```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
import scipy.interpolate as ip
from scipy import stats
from statistics import mean 
from matplotlib.pyplot import figure
import numpy as np
from sklearn.linear_model import LinearRegression




def read_distribution_degree_csv(fname, n,i):
    timings = pd.read_csv(fname, sep=',')
    degree = timings['degree'].values
    size = timings['frequency'].values
    size = np.log(size)

    if i ==0:
        plt.title("Erdos Renyi degree_distribution (n = " + str(n) + ")")
    else:
        plt.title("Barabasi Albert degree_distribution (n = " + str(n) + ")")
    plt.plot(degree, size ,color = "green")
    plt.xlabel("degree")
    plt.ylabel("degree_frequency")


    plt.show()
    plt.close()

    plt.loglog(degree, size,color = "green")
    if i ==0:
        plt.title("Erdos Renyi degree_distribution log-log plot (n = " + str(n) + ")")
    else:
        plt.title("Barabasi Albert degree_distribution log-log plot (n = " + str(n) + ")")
    plt.xlabel("degree")
    plt.ylabel("degree_frequency")
    plt.show()
    plt.close()

        

read_distribution_degree_csv("erdos_degree1000.csv",1000,0)
read_distribution_degree_csv("erdos_degree10000.csv",10000,0)
read_distribution_degree_csv("erdos_degree100000.csv",100000,0)
read_distribution_degree_csv("barabasi_degree1000.csv",1000,1)
read_distribution_degree_csv("barabasi_degree10000.csv",10000,1)
read_distribution_degree_csv("barabasi_degree100000.csv",100000,1)


import numpy as np

def read_diameter_csv(fname,i):
    timings = pd.read_csv(fname, sep=',')
    
    size = timings['n'].values
    size = np.log(size)
    diameter = timings['avg'].values
    if i == 0:
        plt.title("Erdos Renyi - diameter Lin-log plot")
        name = "Erdos Renyi"
    else:
        plt.title("Barabasi Albert - diameter Lin-log plot")
        name = "Barabasi Albert"
        
    plt.plot(size, diameter, 'ro', label=name)
    learner = LinearRegression()
    learner.fit(size.reshape(-1,1),degree.reshape(-1,1))
    degree_predict = learner.predict(size.reshape(-1,1)) 
    plt.loglog(size,degree_predict, label="y = {}x+{}".format(learner.coef_[0][0] , learner.intercept_[0] ,color = "r"))
    
    plt.xlabel("Vertice size")
    plt.ylabel("Average diameter")

    plt.legend()
    plt.show()
    plt.close()

read_diameter_csv("erdos_diameter.csv",0)        
read_diameter_csv("barabasi_diameter.csv",1)
```
